### Creating a scorecard

```python
df_ref_categories = pd.DataFrame(ref_categories, columns = ['Feature name'])
df_ref_categories['Coefficients'] = 0
df_ref_categories['p_values'] = np.nan
df_ref_categories
```

```python
df_scorecard = pd.concat([summary_table, df_ref_categories])
df_scorecard.reset_index(drop=True, inplace=True)
df_scorecard
```

```python
# Concatenate summary_table and df_ref_categories
df_scorecard = pd.concat([summary_table, df_ref_categories], ignore_index=True)
# Extract original feature names
df_scorecard['Original feature name'] = df_scorecard['Feature name'].str.split(':').str[0]
df_scorecard.reset_index(drop=True, inplace=True)
df_scorecard
```

```python
MIN_SCORE, MAX_SCORE = 300, 850
```

```python
df_scorecard.groupby('Original feature name')['Coefficients'].min()
```

```python
min_sum_coef = df_scorecard.groupby('Original feature name')['Coefficients'].min().sum()
min_sum_coef
```

```python
df_scorecard.groupby('Original feature name')['Coefficients'].max()
```

```python
max_sum_coef = df_scorecard.groupby('Original feature name')['Coefficients'].max().sum()
max_sum_coef
```

\begin{equation}
variable\_score =variable\_coef * \frac { (max\_score - min\_score) } { ( max\_sum\_coef -  min\_sum\_coef ) } 
\end{equation}

```python
df_scorecard['Score - Calculation'] = df_scorecard['Coefficients'] * (MAX_SCORE - MIN_SCORE) / (max_sum_coef - min_sum_coef)
df_scorecard
```

```python
df_scorecard.groupby('Original feature name')['Score - Calculation'].min().sum()
```

\begin{equation}
intercept\_score = \frac { (max\_score - min\_score) } { ( max\_sum\_coef -  min\_sum\_coef ) } * ( intercept\_coef - min\_score ) + min\_score
\end{equation}

```python
df_scorecard.at[0, 'Score - Calculation'] = ((df_scorecard['Coefficients'][0] - min_sum_coef) / (max_sum_coef - min_sum_coef)) * (MAX_SCORE - MIN_SCORE) + MIN_SCORE
df_scorecard
```

```python
df_scorecard['Score - Preliminary'] = df_scorecard['Score - Calculation'].round() 
df_scorecard
```

```python
min_sum_score_pre = df_scorecard.groupby('Original feature name')['Score - Preliminary'].min().sum()
min_sum_score_pre
```

```python
max_sum_score_pre = df_scorecard.groupby('Original feature name')['Score - Preliminary'].max(). sum()
max_sum_score_pre
```

```python
df_scorecard['Difference'] = df_scorecard['Score - Preliminary'] - df_scorecard['Score - Calculation']
df_scorecard
```

```python
df_scorecard[df_scorecard['Difference']>0.45]
```

```python
df_scorecard['Score - Final'] = df_scorecard['Score - Preliminary']
df_scorecard.at[81, 'Score - Final'] = df_scorecard.at[81, 'Score - Final'] - 1
df_scorecard.at[1, 'Score - Final'] = df_scorecard.at[1, 'Score - Final'] - 1
```

```python
min_sum_score_final = df_scorecard.groupby('Original feature name')['Score - Final'].min().sum()
min_sum_score_final
```

```python
max_sum_score_final = df_scorecard.groupby('Original feature name')['Score - Final'].max(). sum()
max_sum_score_final
```
